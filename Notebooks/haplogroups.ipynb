```python
import re
hgs = 'A B C D F G H HV HV0 HV1 HV2 I J J1 JT K L0 L01 L1 L2 L3 L4 L5 L6 L7 M N N1 N2 Q R R0 R0a R2 R9 T U V W X'.split()
def mapHGs(hg):
    if re.match('U\d', hg): return 'U'
    if re.match('J2', hg): return 'J'
    if re.match('M\d+', hg): return 'M'
    if re.match('T\d', hg): return 'T'
    if re.match('H\d+', hg): return 'H'
    if hg.startswith('R0a'): return 'R0a'
    if re.match('K\d', hg): return 'K'
    return re.match('[A-Z]+\d*', hg).group()

df = pd.read_csv('../HaplogroupAnalysis/MTHaplo/haplogroups.txt', sep='\t')
yhgs = [mapHGs(hg)  for hg in df.Haplogroup]
counts = [yhgs.count(hg) for hg in hgs]
total = sum(counts)
print('\t'.join(["%.4f"%(c/total) for c in counts]))
df['yhgs'] = yhgs
cols = ['SampleID', 'Haplogroup', 'yhgs', 'Rank', 'Quality', 'Range']
df[cols]
```

## Saudi Arabian Y-Chromosome diversity and its relationship with nearby regions
Saudi Arabian Y-Chromosome diversity and its relationship with nearby regions, Abu-Amero et al
__[Link](https://bmcgenomdata.biomedcentral.com/articles/10.1186/1471-2156-10-59)__

```python
from collections import defaultdict
y = pd.read_excel("../HaplogroupAnalysis/HaplogroupsContext/Y-chr/arabiaY_Yemen.xlsx")
y = y.set_index('Haplogroup').transpose().fillna(0).iloc[1:,:-2]

## fixing duplicate columns (problem in original paper)
## https://bmcgenomdata.biomedcentral.com/articles/10.1186/1471-2156-10-59/tables/1
duplicateCols = ['L1', 'L2*', 'L3*', 'Q1a3*']
for dupCol in duplicateCols:
    y[f'__{dupCol}__'] = y[[dupCol]].sum(axis=1)

keepCols = [col for col in y.columns.values if not col in duplicateCols]
y = y[keepCols]

y.columns = [col.strip('_') for col in y.columns.values]
scols = sorted(y.sum(axis=0).items(), key=lambda x:x[1], reverse=True)
scols = list(list(zip(*scols))[0])
y1 = y[scols[:9]]
y1['Misc'] = 100-y1.sum(axis=1)
y1.sum(axis=1)
fig, ax = plt.subplots(figsize=(16,12))
y1.plot(kind='bar', width=.8, stacked=True, ax=ax)
fig.savefig('../Results/ychrStats.svg')
fig.savefig('../Results/ychrStats.pdf')
```

```python
lumping = {'Ethiopia': 'East Africa', 'Eritrea': 'East Africa','Sudan': 'East Africa',
           'Morocco': 'North Africa', 'Algeria': 'North Africa', 'Egypt': 'North Africa',
           'Georgia': "Caucasus", 'Armenia': "Caucasus", 'Azerbaijan': "Caucasus"}
reichY['Country2'] = [lumping.get(cntry, cntry) for cntry in reichY.Country]
```

```python
countries = []
countryStats = []
for country, ry in reichY.groupby("Country2"):
    if len(ry)>8:
        countryStats.append(Counter([str(code)[:2] for code in ry.Y5]))
        countries.append(country)
#countries
```

```python
        
ychr = pd.DataFrame(countryStats, index=countries)  
ychr = ychr.loc[['Ukraine', 'North Africa', 'East Africa', 'Turkey', 'Caucasus', 'Bulgaria', 'Turkmenistan', 'Israel', 'India', 'Lebanon', 'Pakistan','Iran', 'Jordan']]
ychr = ychr.div(ychr.sum(axis=1), axis=0)

scols = sorted(ychr.sum(axis=0).items(), key=lambda x:x[1], reverse=True)
scols = list(list(zip(*scols))[0])
y1 = ychr[scols[:9]]
y1['Misc'] = 1-y1.sum(axis=1)
y1.sum(axis=1)
fig, ax = plt.subplots(figsize=(16,12))
y1.plot(kind='bar', width=.8, stacked=True, ax=ax)
```
