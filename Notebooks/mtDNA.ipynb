```python
import pandas as pd
import re
from collections import Counter

df = pd.read_csv("../HaplogroupAnalysis/MTHaplo/haplogroups.txt", sep='\t')
```

```python
df
```

```python
Counter(df.Haplogroup)
```

```python
hgs = Counter([re.match('([A-Z])+', haplogroup).group(0) for haplogroup in df.Haplogroup])
labels, values = zip(*hgs.items())
labels, values
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = [8,8]
_=plt.pie(values, labels=labels, autopct='%.00f%%')
```
