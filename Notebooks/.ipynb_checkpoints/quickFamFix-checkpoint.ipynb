```python
## actual fam file update (encoding population into FID (first) column)
famfile = "/home/ahenschel/Dropbox/Yemen/YemenGenomeAnalysis/AdmixTools_Yemen_HO_OOA_geno05/yemen_reichHO_mind0.01_geno_0.05.LD.QC.ooa.ee.fam"
fambak = famfile + '_bak'
os.system(f'mv {famfile} {fambak}')
```

```python
fam = pd.read_csv(fambak, sep=' ',header=None)
fam.head()
```

```python
popIds = []
unknowns = []
for iid in fam[1]:
    iid = str(iid)
    if iid.startswith('urn:wtsi:'): 
        if '3577STDY' in iid: iid = iid.split('_')[-1]
        else: iid = iid.split(':')[-1]
    if iid not in meta:
        popIds.append('unknown')
        unknowns.append(iid)
    else:
        popIds.append(meta[iid])
fam[0] = popIds
```
