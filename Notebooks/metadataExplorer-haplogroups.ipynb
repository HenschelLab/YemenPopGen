```python
def readReich(reichset, contemporaryOnly=False, QC=False, addRegion=False, popmerge=False):
    if reichset=='1240K':
        columns2keep = [1,9, 12, 14, 15, 16]
    elif reichset=='HO':
        columns2keep = [1, 5, 7, 9, 10, 11]
    columnNames = 'Id Date Group_Label Country Lat Long'.split()
    reich=pd.read_csv(f"../Reich/v44.3_{reichset}_public.anno", sep='\t')
    reich = reich.iloc[:,columns2keep]
    reich.columns = columnNames

    if QC:
        reich = reich[~reich.Group_Label.str.startswith("Ignore_")]
        reich = reich[~reich.Group_Label.str.endswith("_outlier")]
    if contemporaryOnly:
        reich = reich[reich.Date==0]
    if addRegion:
        from pycountry_convert import country_alpha2_to_continent_code, country_name_to_country_alpha2
        reich["Region"] = [country2continent(country) for country in reich.Country]
        reich = reich.dropna(subset=['Region']) # drop few entries where we couldnt find the continent
    if popmerge=='global':
        reich.loc[reich.Country=='Yemen', 'Group_Label'] = 'Yemen'   
        ## TODO: combine also other populations based on HC     
    return reich
reich = readReich('1240K', QC=True)
Counter(reich.Country).most_common()
```

## Measuring Geographic distance
This is to identify populations that matter most. Admixture gets swamped if too many populations are included


```python
from math import sqrt, sin, cos, asin, radians
def latlongdist(lat1 ,lon1, lat2=None, lon2=None):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    ## Haversine formula for distance to central Yemen
    if lat2 is None:
        lon2 = reichYemen.Long.astype(float).mean()
        lat2 = reichYemen.Lat.astype(float).mean()

    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])
    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    # Radius of earth in kilometers is 6371
    km = 6371* c
    return km

kouraLat, kouraLong = 34.2992, 35.814            
reichLL.loc[:,"Distance2Yemen"] = [latlongdist(float(row.Lat), float(row.Long)) for i, row in reichLL.iterrows()]
#reich.loc[:,"Distance2Koura"] = [latlongdist(float(row.Lat), float(row.Long), kouraLat, kouraLong) for i, row in reich.iterrows()]
if saveReichLL:
    print(f'Metadata/reichLL_{reichset}.csv', len(reichLL))
    reichLL.to_csv(f'Metadata/reichLL_{reichset}.csv')
```

```python
allPops = Counter(reich.Group_Label)


def filterByGeoDistance(distance=3000):
    nearPops = Counter(reichLL[reichLL.Distance2Yemen < distance].Group_Label)
    vids = []
    for nearPop, count in nearPops.items():
        if 2*count > allPops[nearPop]: # at least 50% of the population are captured with the geographic constraint
            popDF = reich[reich.Group_Label==nearPop]
            vids += list(popDF['Version ID'])
    return vids

def filterByGeoDistance2(distance=3000):
    nearPops = Counter(reichLL[reichLL.Distance2Yemen < distance].Group_Label)
    return [nearPop for nearPop, count in nearPops.items() if 2*count > allPops[nearPop]]

filterByGeoDistance2(1000)
```

```python
fam = pd.read_csv('yemen_clean_reichHO4.fam', sep=' ', header=None)

def saveKeeperFile4Plink(distance=3000, outfile = 'filterpops2'):
    keepers = filterByGeoDistance(distance=distance)
    ## with fam-id and sample-id
    with open(f'{outfile}_{distance}.txt', 'w') as out:
        for idx, row in fam.iterrows(): ## keep also our Yemen samples (not in Reich dataset)
            if row[1] in keepers or '3577STDY' in row[1]:
                print(row[0], row[1], file=out)

for distance in [1000, 2000, 3000, 4000, 5000]:
    saveKeeperFile4Plink(distance=distance)
    
```
